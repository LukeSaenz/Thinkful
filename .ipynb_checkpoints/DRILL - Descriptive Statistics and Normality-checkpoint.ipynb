```python
# Generate a logarithmic series distribution with n=100 and p=0.5.
random_log = np.random.logseries(0.5, 100)

# Plot a histogram.
plt.hist(random_log)

# Print the histogram.
plt.show()
```

Compute the mean and standard deviation and plot them as vertical lines on the histogram.

```python
# Add a vertical line at one standard deviation above the mean.
plt.axvline(random_log.mean() + random_log.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(random_log.mean()-random_log.std(), color='b', linestyle='dashed', linewidth=2) 

plt.hist(random_log)
plt.show()
```

```python
# Generate a binomial distribution with n=100 and p=0.7.
binomial_dist = np.random.binomial(100, 0.7, 100)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binomial_dist.mean() + binomial_dist.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binomial_dist.mean()-binomial_dist.std(), color='b', linestyle='dashed', linewidth=2) 

# Plot a histogram.
plt.hist(binomial_dist)

# Print the histogram.
plt.show()
```

```python
# Generate the geometric distribution with n=100 and p=0.2.
geo = np.random.geometric(0.2, 100)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(geo.mean() + geo.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(geo.mean()-geo.std(), color='b', linestyle='dashed', linewidth=2) 

# Plot a histogram.
plt.hist(geo)

# Print the histogram.
plt.show()
```

```python
# Generate a power distribution with positive exponent a - 1 with n=100 and a = 0.9.
power = np.random.power(0.9, 100)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(power.mean() + power.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(power.mean()-power.std(), color='b', linestyle='dashed', linewidth=2) 

# Plot a histogram.
plt.hist(power)

# Print the histogram.
plt.show()
```

```python
# Generate a Rayleigh distribution with n=100 and scale = 0.5.
rayleigh = np.random.rayleigh(0.5, 100)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(rayleigh.mean() + rayleigh.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(rayleigh.mean()-rayleigh.std(), color='b', linestyle='dashed', linewidth=2) 

# Plot a histogram.
plt.hist(rayleigh)

# Print the histogram.
plt.show()
```

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
