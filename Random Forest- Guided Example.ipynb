```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Replace the path with the correct path for your data.
y2015 = pd.read_csv(
    'LoanStats3d.csv',
    skipinitialspace=True,
    header=1
)

# Note the warning about dtypes.
```

```python
y2015.head()
```

```python
#Looking at all categorical variables to see how many distinct counts there are

categorical = y2015.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
# Convert ID and Interest Rate to numeric.
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state', 'desc'], 1, inplace=True)
```

```python
y2015.tail()
```

```python
#Remove the two summary rows at the end that don't actually contain any data
y2015 = y2015[:-2]
```

```python
pd.get_dummies(y2015)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
#drop NA columns
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=14)
pca.fit(X)
```

```python
print(pca.explained_variance_)
```

```python
X_pca = pca.transform(X)
```

```python
cross_val_score(rfc, X_pca, Y, cv=10)

#Looks like 14 components is as low as I can go while maintaining a 90% average across all 10 folds
```
